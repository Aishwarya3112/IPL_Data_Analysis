```python
import pandas as pd
deliveries = pd.read_csv("/content/deliveries.csv")
```

```python
deliveries.head()
```

```python
deliveries.shape
```

```python
deliveries.info()
```

```python
top_runs=deliveries.groupby("batsman")["batsman_runs"].sum().sort_values(ascending=False).head(10)
top_runs
```

```python
deliveries[deliveries["batsman_runs"] == 6]["batsman"].value_counts()
```

```python
top_wickets=deliveries[deliveries["player_dismissed"].notna()]["bowler"].value_counts().head(10)
top_wickets
```

```python
deliveries.groupby("bowler")["total_runs"].sum().sort_values(ascending=False)
```

```python
toss_decision = matches["toss_decision"].value_counts()
toss_decision
```

```python
matches_per_season = matches["season"].value_counts().sort_index()
matches_per_season
```

```python
import matplotlib.pyplot as plt
team_wins.plot(kind="bar", figsize=(10,5))
plt.title("Matches Won By Each IPL Team")
plt.xlabel("Team")
plt.ylabel("Number of wins")
plt.show()
```

```python
import matplotlib.pyplot as plt
top_runs.plot(kind="bar", figsize=(10,5))
plt.xticks(rotation=45)
plt.title("Top 10 Run scorers in IPL")
plt.xlabel("Batsman")
plt.ylabel("Number of runs")
plt.show()
```

```python
import matplotlib.pyplot as plt
top_wickets.plot(kind="bar", figsize=(10,5))
plt.xticks(rotation=45)
plt.title("Top 10 Wicket Takers in IPL")
plt.xlabel("Bowler")
plt.ylabel("Number of wickets")
plt.show()
```

```python
import matplotlib.pyplot as plt
toss_decision.plot(kind="pie",autopct="%1.1f%%")
plt.title("Toss Decision in IPL")
plt.show()
```

```python
matches_per_season.plot(kind="line", figsize=(10,5))

plt.title("Matches Played Per Season")
plt.xlabel("Season")
plt.ylabel("Matches")

plt.show()
```
